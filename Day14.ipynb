```python
from collections import Counter


def roll_rock_left(sequence: str):
    counter = Counter(sequence)
    return f"{'O'*counter.get('O', 0)}{'.'*counter.get('.', 0)}"


def tilt_north(pattern: list[str]):
    transpose = ["".join(line[col] for line in pattern) for col, _ in enumerate(pattern[0])]
    newlines = ["#".join([roll_rock_left(t) for t in tl.split("#")]) for tl in transpose]
    # transpose back
    newlines = ["".join(line[col] for line in newlines) for col, _ in enumerate(newlines[0])]
    return newlines


def get_load(pattern: list[str]):
    nlines = len(pattern)
    load = 0
    for lidx, line in enumerate(pattern):
        counter = Counter(line)
        load += counter.get("O", 0) * (nlines-lidx)
    return load


def parse_data(data: str):
    lines = [line for line in data.split("\n") if line]
    return lines
        
```

```python
def get_north_load(pattern: list[str]):
    pattern = tilt(pattern, "N")
    nlines = len(pattern)
    load = 0
    for lidx, line in enumerate(pattern):
        counter = Counter(line)
        load += counter.get("O", 0) * (nlines-lidx)
    return load

def tilt(pattern: list[str], direction: str):
    # transpose
    if direction.upper()[0] in ["N", "S"]:
        pattern = ["".join(line[col] for line in pattern) for col, _ in enumerate(pattern[0])]
    # flip
    if direction.upper()[0] in ["S", "E"]:
        pattern = [line[::-1] for line in pattern]

    pattern = ["#".join([roll_rock_left(p) for p in pl.split("#")]) for pl in pattern]

    # flip back
    if direction.upper()[0] in ["S", "E"]:
        pattern = [line[::-1] for line in pattern]
    # transpose back
    if direction.upper()[0] in ["N", "S"]:
        pattern = ["".join(line[col] for line in pattern) for col, _ in enumerate(pattern[0])]
    return pattern

def spin(pattern: list[str]):
    for dirct in ["N", "W", "S", "E"]:
        pattern = tilt(pattern, dirct)
    return pattern

def load_recorder(pattern: list[str], cycle: int = 1_000_000_000):
    for i in range(cycle):
        pattern = spin(pattern)
        yield get_load(pattern)

def guess_load(pattern: list[str]):
    load = []
    regularity = None
    for l in load_recorder(pattern):
        load.append(l)
        for i in range(2, len(load)//2):
            if load[-i:] == load[-2*i:-i]:
                regularity = load[-1-i:-1]
                break

        if regularity:
            break
    idx = (int(1e9) - len(load)) % len(regularity)
    # print(regularity)
    return regularity[idx]

```
