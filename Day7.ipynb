```python
import re
from functools import total_ordering

class Card:
    order_mapping = {"A": 14, "K": 13, "Q": 12, "J": 11, "T":10}
    def __init__(self, card: str) -> None:
        self.card = card

    def __eq__(self, obj: Card) -> bool:
        return self.index == obj.index

    def __lt__(self, obj: Card) -> bool:
        return self.index < obj.index
    
    def __repr__(self) -> str:
        return f"Card({self.card})"

    @property    
    def index(self) -> int:
        return self.order_mapping.get(self.card, None) or int(self.card)

@total_ordering
class Hand:

    def __init__(self, cards: str, bid: int):
        if len(cards) != 5:
            raise ValueError(f"'{cards}' must have exact 5 cards")

        self.cards = [Card(c) for c in cards]
        self.bid = bid
        self.pattern_point = self._get_pattern_point(cards)
        self._rank = 0

    def __eq__(self, obj: Hand) -> bool:
        return all([c1 == c2 for c1, c2 in zip(self.cards, obj.cards)])

    def __lt__(self, obj: Hand):
        if self.pattern_point != obj.pattern_point:
            return self.pattern_point < obj.pattern_point
        else:
            for c1, c2 in zip(self.cards, obj.cards):
                if c1 != c2:
                    return c1 < c2

    def __repr__(self) -> str:
        return f"Hand({''.join(c.card for c in self.cards)}, {self.bid})"
    
    def _get_pattern_point(self, cards: str) -> int:
        pattern_point = 0

        for c in cards:
            count = len(re.findall(c, cards))
            cards = cards.replace(c, "")
            pattern_point += count**2
            # print(c, cards, count, self.pattern_point)

        return pattern_point

    @property
    def rank(self) -> int:
        return self._rank
    
    @rank.setter
    def rank(self, new) -> None:
        if self._rank:
            raise RuntimeError("Repeating set rank")
        self._rank = new
```

```python
def parse_data(data: str):
    lines = [line for line in data.split("\n") if line]
    hands = [Hand(line.split(" ")[0], int(line.split(" ")[1])) for line in lines]
    return hands
    
```

```python
import numpy as np
def get_total_win(data: str):
    sorted_hands = sorted(parse_data(data))
    return np.sum(np.array([h.bid for h in sorted_hands])*np.arange(1, len(sorted_hands)+1))
```

```python
class Card2(Card):
    order_mapping = {"A": 14, "K": 13, "Q": 12, "J": 1, "T":10}

class Hand2(Hand):

    def __init__(self, cards: str, bid: int):
        super().__init__(cards, bid)

        self.cards = [Card2(c) for c in cards]
        self.pattern_point = self._get_pattern_point(cards)

    def _get_pattern_point(self, cards: str) -> int:
        pattern_point = 0

        nJ = len(re.findall("J", cards))
        cards = cards.replace("J", "")

        counts_list = []
        for c in cards:
            counts_list.append((c, len(re.findall(c, cards))))
            cards = cards.replace(c, "")

        counts_list.sort(key=lambda x: x[1], reverse=True)

        for _, n in counts_list:
            pattern_point += (n + nJ)**2
            nJ = 0

        if nJ > 0:
            pattern_point += nJ**2

        return pattern_point
    
def parse_data2(data: str):
    lines = [line for line in data.split("\n") if line]
    hands = [Hand2(line.split(" ")[0], int(line.split(" ")[1])) for line in lines]
    return hands
    
def get_total_win2(data: str):
    sorted_hands = sorted(parse_data2(data))
    return np.sum(np.array([h.bid for h in sorted_hands])*np.arange(1, len(sorted_hands)+1))
```
