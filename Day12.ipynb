```python
import numpy as np

### Method 1
def symbol_record_to_number(record: str):
    number_record = []
    continuation = 0
    for i, c in enumerate(record):
        if c == "#":
            continuation += 1
            if i == len(record) - 1:
                number_record.append(continuation)
        elif c == "." and continuation > 0:
            number_record.append(continuation)
            continuation = 0
    return number_record

def get_digit_in_binary(number: int, digit: int):
    binary = bin(number)
    return int(binary[-1-digit]) if digit < len(binary) - 2 else 0

def generate_possible_record(partial_record: str):
    unknown_idx = [i for i, c in enumerate(partial_record) if c == "?"]
    for i in range(2**len(unknown_idx)):
        replacement = {}
        for iidx, idx in enumerate(unknown_idx):
            replacement.update({idx: "#" if get_digit_in_binary(i, iidx) else "."})
        record = "".join([replacement.get(idx) if idx in unknown_idx else c for idx, c in enumerate(partial_record)])
        yield record

def get_number_possible_arrangement(symbol_record: str, digit_record: list[int]):
    noption = 0
    for possible_record in generate_possible_record(symbol_record):
        if symbol_record_to_number(possible_record) == digit_record:
            noption += 1
    
    return noption

def get_total_number_possible_arrangement(break_symbol_records: list[str], digit_records: list[list[int]]):
    noption = 0
    for break_symbol_record, digit_record in zip(break_symbol_records, digit_records):
        noption += get_number_possible_arrangement(break_symbol_record, digit_record)
    return noption


### Method 2
def validpiece(piece: str):
    return all([p in ["?", "#"] for p in piece])

def get_possible_start_positions(symbol_record: str, digit_record: str):

    possible_start_positions = []
    for idx, digit in enumerate(digit_record):
        predigits = int(np.sum(digit_record[:idx])) + idx
        postdigits = int(np.sum(digit_record[idx+1:])) + len(digit_record[idx+1:])

        possible_start = []
        for position in range(predigits, len(symbol_record)-postdigits):
            if position + digit > len(symbol_record):
                break
            if validpiece(symbol_record[position:position+digit]):
                if (position == 0 or symbol_record[position-1] in [".", "?"]):
                    if (position+digit == len(symbol_record) or symbol_record[position+digit] in [".", "?"]):
                        possible_start.append(position)
        possible_start_positions.append(possible_start)
    return possible_start_positions

def generate_compositions(possible_start_positions: list[list[int]]):
    shape = [len(possible_start) for possible_start in possible_start_positions]
    accumshape = [1]
    for s in shape:
        accumshape.append(accumshape[-1]*s)

    for i in range(accumshape[-1]):
        composition = []
        for s, a, possible_start in zip(shape, accumshape[:-1], possible_start_positions):
            composition.append(possible_start[(i//a)%s])

        yield composition

def valid_composition(symbol_record: str, digit_record: list[int], composition: list[int]):
    for l, n in zip(composition[:-1], composition[1:]):
        if l >= n:
            return False

    replaced_record = ["."]*len(symbol_record)
    for i, d in zip(composition, digit_record):
        for idx in range(i, i+d):
            replaced_record[idx] = "#"

    replaced_record = "".join(replaced_record)

    for c, r in zip(symbol_record, replaced_record):
        if c == "#" and r != "#":
            return False

    return symbol_record_to_number(replaced_record) == digit_record

def get_number_possible_arrangement_by_digit(symbol_record: str, digit_record: list[int]):
    if all([c == "?" for c in symbol_record]):
        ndots = len(symbol_record) - int(np.sum(digit_record))
        noption = 1
        for i, _ in enumerate(digit_record):
            noption *= (ndots+1-i) / (i+1) # C^{ndots+1}_{#digit}
        return int(noption)

    noption = 0
    for comp in generate_compositions(get_possible_start_positions(symbol_record, digit_record)):
        if valid_composition(symbol_record, digit_record, comp):
            noption += 1
    return noption

def get_total_number_possible_arrangement_by_digit(break_symbol_records: list[str], digit_records: list[list[int]]):
    noption = 0
    for break_symbol_record, digit_record in zip(break_symbol_records, digit_records):
        noption += get_number_possible_arrangement_by_digit(break_symbol_record, digit_record)
        # ground_truth = get_number_possible_arrangement(break_symbol_record, digit_record)
        # new_method = get_number_possible_arrangement_by_digit(break_symbol_record, digit_record)
        # if ground_truth != new_method:
        #     print(break_symbol_record, digit_record, ground_truth, new_method)
        #     raise RuntimeError
    return noption


### Method 3
def get_number_possible_arrangement_recursive(symbol_record: str, digit_record: list[int]): #, cache: dict[tuple[str, list[str]], int]=None):

    checked = {}
    uncheck = [(symbol_record, tuple(digit_record))]
    # noption = 0

    while uncheck:

        # cur_record, cur_digits = uncheck.pop()
        cur_record, cur_digits = uncheck[-1]

        if not cur_digits:
            if "#" in cur_record:
                checked.update({(cur_record, cur_digits): 0})
                uncheck.pop()
            else:
                checked.update({(cur_record, cur_digits): 1})
                uncheck.pop()
                # noption += 1

        elif len(cur_record) < cur_digits[0]:
            checked.update({(cur_record, cur_digits): 0})
            uncheck.pop()

        else:

            cur_num = 0
            add_checked = True
            if cur_record[0] in ["?", "#"]:
                if len(cur_record) >= cur_digits[0] and "." not in cur_record[:cur_digits[0]] and (len(cur_record) == cur_digits[0] or cur_record[cur_digits[0]] != "#"):
                    if checked.get((cur_record[cur_digits[0]+1:], cur_digits[1:])) is not None:
                        cur_num += checked.get((cur_record[cur_digits[0]+1:], cur_digits[1:]))
                        
                    else:
                        uncheck.append((cur_record[cur_digits[0]+1:], cur_digits[1:]))
                        add_checked = False

            if cur_record[0] in ["?", "."]:                
                if checked.get((cur_record[1:], cur_digits)) is not None:
                    cur_num += checked.get((cur_record[1:], cur_digits))
                    
                else:
                    uncheck.append((cur_record[1:], cur_digits))
                    add_checked = False

            if add_checked:
                checked.update({(cur_record, cur_digits): cur_num})
                uncheck.pop()

    return checked.get((symbol_record, tuple(digit_record)))

def get_total_number_possible_arrangement_recursive(symbol_records: list[str], digit_records: list[list[int]]):
    noption = 0
    for symbol_record, digit_record in zip(symbol_records, digit_records):
        noption += get_number_possible_arrangement_recursive(symbol_record, digit_record)
    return noption

def get_total_number_possible_arrangement_recursive_unfold(symbol_records: list[str], digit_records: list[list[int]]):
    noption = 0
    for idx, (symbol_record, digit_record) in enumerate(zip(symbol_records, digit_records)):
        noption += get_number_possible_arrangement_recursive("?".join([symbol_record]*5), digit_record*5)
    return noption

def parse_data(data: str):
    lines = [line for line in data.split("\n") if line]
    symbol_records = []
    digit_records = []
    for line in lines:
        symbol_record, digit_record = line.split(" ")
        symbol_records.append(symbol_record)
        digit_record = [int(c) for c in digit_record.split(",")]
        digit_records.append(digit_record)
    return symbol_records, digit_records
```
