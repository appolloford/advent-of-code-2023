```python
class EngineChar:
    def __init__(self, char: str, row: int, col: int):
        self.char = char
        self.row = row
        self.col = col
        self.adjacent_number = []

    def __repr__(self) -> str:
        return f"EngineChar({self.char}, {self.row}, {self.col})"
    
    def add_adjacent_number(self, num: int):
        self.adjacent_number.append(num)

    @property
    def ratio(self) -> int:
        if len(self.adjacent_number) > 2:
            raise ValueError(f"Too many numbers: {self.add_adjacent_number}")
        if self.char == "*" and len(self.adjacent_number) == 2:
            return self.adjacent_number[0] * self.adjacent_number[1]
        else:
            print("Skip...", self.char, self.adjacent_number)
            return 0

class EngineNumber:
    def __init__(self, value: int, row: int, left: int, right: int):
        self.value = value
        self.row = row
        self.left = left
        self.right = right

    def isadjacent(self, echar: EngineChar):
        if echar.row >= self.row-1 and echar.row <= self.row+1:
            if echar.col >= self.left-1 and echar.col <= self.right+1:
                return True
        return False
    
    def __repr__(self) -> str:
        return f"EngineNumber({self.value}, {self.row}, {self.left}, {self.right})"
    
def parse_row(line: str, row: int):
    enginechar_in_line = []
    enginenumber_in_line = []
    tmpnum = ""
    for col, c in enumerate(line):
        if c.isdigit():
            tmpnum += c
            if col == len(line) - 1:
                enginenumber_in_line.append(EngineNumber(int(tmpnum), row, col-len(tmpnum), col-1))
                tmpnum = ""
        elif not c.isalnum():
            if c != ".":
                enginechar_in_line.append(EngineChar(c, row, col))
            if tmpnum:
                enginenumber_in_line.append(EngineNumber(int(tmpnum), row, col-len(tmpnum), col-1))
                tmpnum = ""
        else:
            raise ValueError(f"Unexpected value: {c}")
    return enginechar_in_line, enginenumber_in_line
        
def parse_data(input: str):
    lines = [s for s in input.split("\n") if s]
    enginenumber_in_data = []
    enginechar_in_data = []
    for row, line in enumerate(lines):
        char_in_line, num_in_line = parse_row(line, row)
        enginechar_in_data.extend(char_in_line)
        enginenumber_in_data.extend(num_in_line)
    return enginechar_in_data, enginenumber_in_data
```

```python
number_with_adjacent = []
for num in nums:
    for char in chars:
        if num.isadjacent(char):
            number_with_adjacent.append(num.value)
print(number_with_adjacent)
import numpy as np
print(np.sum(number_with_adjacent))
```
