```python
def get_horizontal_mirror(pattern: list[str]):
    nrow = len(pattern)
    for idx, line in enumerate(pattern):
        if (nrow - idx) % 2:
            continue
        
        if pattern[idx:(idx+nrow)//2] == pattern[:(idx+nrow)//2-1:-1]:
            return (idx+nrow)//2

    reverse = pattern[::-1]
    for idx, line in enumerate(reverse):
        if (nrow - idx) % 2:
            continue
        
        if reverse[idx:(idx+nrow)//2] == reverse[:(idx+nrow)//2-1:-1]:
            return nrow - (idx+nrow)//2
    return -1

def get_vertical_mirror(pattern: list[str]):
    transpose = ["".join(line[col] for line in pattern) for col, _ in enumerate(pattern[0])]
    return get_horizontal_mirror(transpose)

def get_mirror_number(pattern: list[str]):
    horizontal = get_horizontal_mirror(pattern)
    vertical = get_vertical_mirror(pattern)
    # print(horizontal, vertical)
    if horizontal > 0 and vertical > 0:
        raise RuntimeError("Something wrong, two mirrors")
    elif horizontal > 0:
        return 100 * horizontal
    elif vertical > 0:
        return vertical
    else:
        raise RuntimeError("No mirror?")
    
def get_total_number(patterns: list[list[str]]):
    num = 0
    for pattern in patterns:
        num += get_mirror_number(pattern)
    return num

def parse_data(data: str):
    lines = data.split("\n")
    pattern, patterns = [], []
    for line in lines:
        if not line:
            if pattern:
                patterns.append(pattern)
            pattern = []
        else:
            pattern.append(line)

    return patterns
```

```python
def get_sites_replace_by_row(pattern: list[str]):
    sites = []
    for lidx, line in enumerate(pattern):
        hidx = set([i for i, c in enumerate(line) if c == "#"])
        for nidx, nline in enumerate(pattern[lidx:]):
            nhidx = set([i for i, c in enumerate(nline) if c == "#"])
            if len(hidx.difference(nhidx)) == 1:
                sites.append((lidx, nidx, *hidx.difference(nhidx)))
            elif len(nhidx.difference(hidx)) == 1:
                sites.append((nidx + lidx, lidx, *nhidx.difference(hidx)))
    return sites

def get_all_horizontal_mirror(pattern: list[str]):
    mirrors = []
    nrow = len(pattern)
    for idx, line in enumerate(pattern):
        if (nrow - idx) % 2:
            continue
        
        if pattern[idx:(idx+nrow)//2] == pattern[:(idx+nrow)//2-1:-1]:
            mirrors.append((idx+nrow)//2)

    reverse = pattern[::-1]
    for idx, line in enumerate(reverse):
        if (nrow - idx) % 2:
            continue
        
        if reverse[idx:(idx+nrow)//2] == reverse[:(idx+nrow)//2-1:-1]:
            mirrors.append(nrow - (idx+nrow)//2)
    return mirrors

def get_new_horizontal_mirror(pattern: list[str]):
    sites = get_sites_replace_by_row(pattern)
    all_horizontal_mirror_place = []
    for l, n, i in sites:
        old_mirror = get_horizontal_mirror(pattern)
        new_pattern = [f"{line[:i]}.{line[i+1:]}" if idx == l else line for idx, line in enumerate(pattern)]
        all_horizontal_mirror_place = get_all_horizontal_mirror(new_pattern)
        if old_mirror in all_horizontal_mirror_place:
            all_horizontal_mirror_place.remove(old_mirror)
        
        if all_horizontal_mirror_place:
            break
    return all_horizontal_mirror_place

def get_new_vertical_mirror(pattern: list[str]):
    transpose = ["".join(line[col] for line in pattern) for col, _ in enumerate(pattern[0])]
    return get_new_horizontal_mirror(transpose)

def get_new_mirror_number(pattern: list[str]):
    horizontal = get_new_horizontal_mirror(pattern)
    vertical = get_new_vertical_mirror(pattern)
    if len(horizontal) > 1 or len(vertical) > 1 or (len(horizontal) and len(vertical)):
        raise RuntimeError("Something wrong, too many mirrors")
    elif horizontal:
        return 100 * horizontal[0]
    elif vertical:
        return vertical[0]
    else:
        raise RuntimeError("No mirror?")
    
def get_new_total_number(patterns: list[list[str]]):
    num = 0
    for pattern in patterns:
        num += get_new_mirror_number(pattern)
    return num
```
