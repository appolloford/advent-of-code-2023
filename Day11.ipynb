```python
import numpy as np

def expand_data(lines):
    extend_col = []
    for col, _ in enumerate(lines[0]):
        if all([line[col] == "." for line in lines]):
            extend_col.append(col)

    newlines = []
    for line in lines:
        newline = "".join([".." if col in extend_col else c for col, c in enumerate(line)])
        if all([ele == "." for ele in newline]):
            newlines.append(newline)
        newlines.append(newline)

    return newlines

def get_galaxies_position(lines):
    positions = []
    for row, line in enumerate(lines):
        for col, c in enumerate(line):
            if c == "#":
                positions.append((row, col))

    return positions

def get_total_distance(positions):
    distance = 0
    for idx, (row, col) in enumerate(positions):
        for pairidx, (pairrow, paircol) in enumerate(positions[idx+1:]):
            distance += np.abs(pairrow-row) + np.abs(paircol-col)
    return distance

def parse_data(data: str):
    lines = [line for line in data.split("\n") if line]
    return lines
```

```python
lines = parse_data(inputdata)
lines = expand_data(lines)
positions = get_galaxies_position(lines)
get_total_distance(positions)
```

```python
def get_extend_col(lines):
    extend_col = []
    for col, _ in enumerate(lines[0]):
        if all([line[col] == "." for line in lines]):
            extend_col.append(col)
    return extend_col

def get_extend_row(lines):
    extend_row = []
    for row, line in enumerate(lines):
        if all([ele == "." for ele in line]):
            extend_row.append(row)
    return extend_row

def get_total_distance2(positions, extend_row, extend_col, expansion_rate=1):
    distance = 0
    for idx, (row, col) in enumerate(positions):
        for pairidx, (pairrow, paircol) in enumerate(positions[idx+1:]):
            minrow, maxrow = sorted([row, pairrow])
            mincol, maxcol = sorted([col, paircol])
            for r in range(minrow, maxrow):
                if r in extend_row:
                    distance += expansion_rate
                distance += 1
            for c in range(mincol, maxcol):
                if c in extend_col:
                    distance += expansion_rate
                distance += 1
    return distance
```

```python
lines = parse_data(inputdata)
extend_row = get_extend_row(lines)
extend_col = get_extend_col(lines)
positions = get_galaxies_position(lines)
get_total_distance2(positions, extend_row, extend_col, 999999)
```
