```python
def parse_data(data: str):
    lines = [line for line in data.split("\n") if line]
    instructions = lines[0]
    nodes_left, nodes_right = dict(), dict()
    for line in lines[1:]:
        line = line.replace("= (", "").replace(",", "").replace(")", "")
        start, left, right = line.split(" ")
        nodes_left[start.strip()] = left.strip()
        nodes_right[start.strip()] = right.strip()

    return instructions, nodes_left, nodes_right

def run(instructions: str, nodes_left: dict[str, str], nodes_right: dict[str, str]):
    current = "AAA"
    step = 0

    while 1:
        step_insturction = instructions[step%len(instructions)]
        step += 1
        if step_insturction == "L":
            current = nodes_left.get(current)
        elif step_insturction == "R":
            current = nodes_right.get(current)
        else:
            raise ValueError(f"Unknown step: {step}")

        yield current, step

def count_steps(data: str):
    for cur, i in run(*parse_data(data)):
        if cur == "ZZZ":
            print("total steps", i)
            break
```

```python
def run3(starting_node:str, instructions: str, nodes_left: dict[str, str], nodes_right: dict[str, str]):
    current = starting_node
    step = 0

    while 1:
        step_insturction = instructions[step%len(instructions)]
        step += 1
        if step_insturction == "L":
            current = nodes_left.get(current)
        elif step_insturction == "R":
            current = nodes_right.get(current)
        else:
            raise ValueError(f"Unknown step: {step}")
        # print(current, step)

        yield current, step

def count_steps_each(data: str):
    starting_nodes, instructions, nodes_left, nodes_right = parse_data2(data)
    steps = []
    for start in starting_nodes:
        for cur, i in run3(start, instructions, nodes_left, nodes_right):
            if cur.endswith("Z"):
                print(f"minimal steps for start {start}, {i}")
                break
        steps.append(i)
    return steps

def find_factors(num: int):
    factors = []
    for i in range(1, num):
        if i*i > num:
            break
        if num % i == 0:
            factors.extend([i, num//i])
    return sorted(factors)
```
